```python
import datetime as dtm      # Python's standard date and time library
import dateutil as dutil    # Powerful datetime extensions
import glob
from math import *
from matplotlib.dates import num2date
import matplotlib.pyplot as plt
%matplotlib inline
from netCDF4 import Dataset
import numpy as np
import pytz                 # Timezone extensions
import seaborn as sns
import scipy.io
```

```python
def get_dataframe(file,u_key='u',v_key='v',w_key='w',time_key='time',temp_key='temp',depth_index=26):
    '''
    INPUT:
    file: filename string
    u_key: key string for eastward velocity, default is 'u'
    v_key: key string for northward velocity, default is 'v'
    w_key: key string for vertical velocity, default is 'w'
    time_key: key string for time, default is 'time'
    temp_key: key string for temperature, default is 'temp'
    depth_index: integer defining depth slice. Default is 26, depth[26]=170.6 m.
    
    OUTPUT:
    df: dataframe with columns time,u,v,w,temp
    '''
    data = Dataset(file)
    time = data.variables['time']
    uu = data.variables['u']
    vv = data.variables['v']
    ww = data.variables['w']
  

def append_dataframe(df, df2):
    '''INPUT
    df: dataframe to append to.
    df2: dataframe to be appended.
    
    OUTPUT:
    Appended dataframes
    '''
    return(df.append(df2, ignore_index=True))
    
def ONC_to_pytime(time):
    '''
    Transform dates into something python can use.
    Returns python datetimes array from time.
    '''
    date1 = dtm.date(1, 1, 1)
    date2 = dtm.date(1970,1,1)
    timedelta1 = date2 - date1
    python_datetimes = np.array([(dtm.datetime.fromordinal(int(tt)) 
                                  + dtm.timedelta(days=tt%1)
                                  + timedelta1 
                                  + dtm.timedelta(days=1)) for tt in time])
    return(python_datetimes)


#All filenames for May-Sep 
filenames = glob.glob('ADCP75kHzPOD2/2014/BarkleyCanyon_UpperSlope_ADCP75kHz_2014*.nc')

df_all = get_dataframe(filenames[0],u_key='u',v_key='v',w_key='w',time_key='time',temp_key='temp',depth_index=26)

for filename in filenames[1:]:
    df = get_dataframe(filename,u_key='u',v_key='v',w_key='w',time_key='time',temp_key='temp',depth_index=26)
    df_all = append_dataframe(df_all, df)
    

    
```

```python
site_name = 'UpperSlope_ADCP_2013-05'
lon = data.variables['longitude']
lat = data.variables['latitude']
depth = data.variables['depth']

bathy = scipy.io.loadmat('Bathymetry/SouthVIgrid.mat')

vigrid = bathy['SouthVIgrid']
lon_bathy = vigrid[0,0][0]
lon_bathy = np.squeeze(np.transpose(lon_bathy))
lat_bathy = vigrid[0,0][1]
lat_bathy = np.squeeze(np.transpose(lat_bathy))
dep_bathy = vigrid[0,0][2]

sns.set_context('talk')
sns.set_style('white')

fig, ax = plt.subplots(1, 2, figsize=(18, 6))
plt.rcParams['contour.negative_linestyle'] = 'solid'

CS = ax[0].contour(lon_bathy,lat_bathy,dep_bathy,
                 (-3000,-2000,-1200,-1000,-500,-300,-200,-100,0),
                 linewidths=(1,1,1,1,1,1,1),
                 colors = 'k')

CS = ax[0].contourf(lon_bathy,lat_bathy,dep_bathy,(0), 
                  colors = ('w',(0.86,0.86,0.86)))

ax[0].plot((lon_bathy[400],lon_bathy[700],lon_bathy[700],lon_bathy[400],lon_bathy[400]),
           (lat_bathy[250],lat_bathy[250],lat_bathy[450],lat_bathy[450],lat_bathy[250]),
           'h-',
           color='OrangeRed',
           linewidth= 4,
           markersize=1)

CS = ax[1].contour(lon_bathy[400:700],lat_bathy[250:500],dep_bathy[250:500,400:700],
                   (-1500,-1000,-800,-600,-400,-300,-200,-100,-50,0),
                   linewidths=(1,1,1,1,1,1,2,1,1,1),
                   colors='k')
ax[1].plot(lon,lat,
           'o',
           color='OrangeRed',
           markersize=14)

ax[1].clabel(CS, inline=1,
             fmt='%1.0f',
             fontsize=15)

ax[0].set_xlabel('Lon', fontsize = 16)
ax[1].set_xlabel('Lon', fontsize = 16)
ax[0].set_ylabel('Lat', fontsize = 16)

```
